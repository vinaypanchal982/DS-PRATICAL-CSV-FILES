```python
import pandas as pd
from scipy import stats

# Load the dataset
df = pd.read_csv("mtcars.csv")

# Check first few rows
print(df.head())

# Separate mpg values based on number of cylinders
mpg_4 = df[df['cyl'] == 4]['mpg']
mpg_6 = df[df['cyl'] == 6]['mpg']
mpg_8 = df[df['cyl'] == 8]['mpg']

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(mpg_4, mpg_6, mpg_8)

# Print results
print("F-statistic:", f_stat)
print("P-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Mean mpg differs across cylinder groups")
else:
    print("No significant difference in mean mpg")
```
